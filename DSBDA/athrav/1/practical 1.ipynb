```python
data={'Name':['Jai','Princi','Gaurav','Anuj','Ravi','Natasha','Riya'],
'Age':[17,17,18,17,18,17,17],
'Gender':['M','F','M','M','M','F','F'],
'Marks':[90,76,'NaN',74,65,'NaN',71]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
c = avg = 0
for ele in df['Marks']:
    if str(ele).isnumeric():  # Ensure 'ele' is numeric
        c += 1
        avg += float(ele)  # Convert to float to avoid type issues

if c != 0:
    avg /= c
else:
    avg = 0  # To avoid ZeroDivisionError
```

```python
df=df.replace(to_replace="NaN", value=avg)
```

```python
df
```

```python
df['Gender'] = df['Gender'].map({'M': 0,'F': 1, }).astype(float)
```

```python
df
```

```python
df = df[df['Marks'] >= 75]
```

```python
df = df.drop(['Age'], axis=1)
```
